```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming 'class' column exists in the DataFrame 'iris2'
classes = iris2['class']
unique_classes = classes.unique()

# Map class labels to colors
colors = {unique_classes[i]: plt.cm.tab10(i / len(unique_classes)) for i in range(len(unique_classes))}

# Creating subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

# First subplot - Color based on class
for class_label in unique_classes:
    idx = classes == class_label
    ax1.plot(iris2['sl'][idx], iris2['sw'][idx], 'o', label=f'Class {class_label}', color=colors[class_label])
ax1.set_title('Scatter Plot 1')
ax1.set_xlabel('Sepal Length')
ax1.set_ylabel('Sepal Width')
ax1.legend()

# Second subplot - Color based on class
for class_label in unique_classes:
    idx = classes == class_label
    ax2.plot(iris2['pl'][idx], iris2['pw'][idx], '^', label=f'Class {class_label}', color=colors[class_label])
ax2.set_title('Scatter Plot 2')
ax2.set_xlabel('Petal Length')
ax2.set_ylabel('Petal Width')
ax2.legend()

# Adjust the layout
plt.tight_layout()

# Display the plot
plt.show()
```

```python
import numpy as np

print("old Shape: ", irispath.shape)

# IQR 
Q1 = np.percentile(irispath['sw'], 25, 
                interpolation = 'midpoint') 

Q3 = np.percentile(irispath['sw'], 75, 
                interpolation = 'midpoint') 
IQR = Q3 - Q1 

# Upper bound 
upper = np.where(irispath['sw'] >= (Q3+1.5*IQR)) 

# Lower bound 
lower = np.where(irispath['sw'] <= (Q1-1.5*IQR)) 

# Removing the Outliers 
irispath.drop(upper[0], inplace = True) 
irispath.drop(lower[0], inplace = True) 

print("New Shape: ", irispath.shape) 

sns.boxplot(x='sw', data=irispath)
```

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler, MinMaxScaler
import pandas as pd

# Load Iris dataset
iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# Add target column for classification
df['species'] = iris.target

# Standardization
scaler_std = StandardScaler()
sepal_length_std = scaler_std.fit_transform(df[['sepal length (cm)']])
sepal_width_std = scaler_std.fit_transform(df[['sepal width (cm)']])
petal_length_std = scaler_std.fit_transform(df[['petal length (cm)']])
petal_width_std = scaler_std.fit_transform(df[['petal width (cm)']])

# Normalization
scaler_minmax = MinMaxScaler()
sepal_length_norm = scaler_minmax.fit_transform(df[['sepal length (cm)']])
sepal_width_norm = scaler_minmax.fit_transform(df[['sepal width (cm)']])
petal_length_norm = scaler_minmax.fit_transform(df[['petal length (cm)']])
petal_width_norm = scaler_minmax.fit_transform(df[['petal width (cm)']])

# Combine into DataFrame
df_standardized = pd.DataFrame({
    'sepal length (cm)_Std': sepal_length_std.flatten(),
    'sepal width (cm)_Std': sepal_width_std.flatten(),
    'petal length (cm)_Std': petal_length_std.flatten(),
    'petal width (cm)_Std': petal_width_std.flatten()
})

df_normalized = pd.DataFrame({
    'sepal length (cm)_Norm': sepal_length_norm.flatten(),
    'sepal width (cm)_Norm': sepal_width_norm.flatten(),
    'petal length (cm)_Norm': petal_length_norm.flatten(),
    'petal width (cm)_Norm': petal_width_norm.flatten()
})
```

```python
import pandas as pd

# Create a DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'HR', 'IT'],
    'Salary': [50000, 60000, 70000, 80000, 75000]
}
df = pd.DataFrame(data)

# Group data by 'Department' and calculate average salary
grouped = df.groupby('Department')['Salary'].mean()
print("Average Salary by Department:")
print(grouped)
```
